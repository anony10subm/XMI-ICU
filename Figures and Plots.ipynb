```python
import pandas as pd
import matplotlib.pyplot as plt
from bumpplot import bumpchart
```

```python
data = {"APACHE (+)":[1,1,1,1],"Albumin (-)":[2,3,4,6],"Tidal Volume (-)":[3,2,2,2], "WBC (+)":[4,8,10,7], 
        "LPM O2 (+)":[5,7,6,3], "Age (+)":[6,9,5,9], "MAP (+)":[7,300,12,300], "Time Since Admission (+)":[301,302,303,11], "Lactate (+)":[8,5,8,5],
        "Systolic BP (-)":[9,4,3,4], "MCHC (-)":[10,300,300,10], "RR (+)":[11,11,300,12], "FiO2 (+)":[12,6,7,8]}
df = pd.DataFrame(data, index=['24 hours','18 hours','12 hours','6 hours'])

plt.figure(figsize=(10, 15))
bumpchart(df, show_rank_axis= False, scatter= True, holes= True,
          line_args= {"linewidth": 5, "alpha": 0.5}, scatter_args= {"s": 100, "alpha": 0.8}) ## bump chart class with nice examples can be found on github

plt.savefig("Time_Shap.png", bbox_inches='tight', dpi=600)
plt.show()
```
